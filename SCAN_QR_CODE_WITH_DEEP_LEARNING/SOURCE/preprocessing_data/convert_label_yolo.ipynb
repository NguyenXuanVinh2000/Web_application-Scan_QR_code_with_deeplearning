```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("label_yolo.csv")
```

```python
df
```

```python
def convert_coordinates(size, box):
    dw = 1.0/size[0]
    dh = 1.0/size[1]
    x = (box[0]+box[1])/2.0
    y = (box[2]+box[3])/2.0
    w = box[1]-box[0]
    h = box[3]-box[2]
    x = x*dw
    w = w*dw
    y = y*dh
    h = h*dh
    return (x,y,w,h)
```

```python
def cover_toyolo():
    for i in range(0,1816):
        fname = df.image[i]
        fname_out = (fname[:-4]+'.txt')
        with open(fname_out, "w") as f:
            width = int(df.Width[i])
            height = int(df.Height[i])
            box = df.QR[i]
            lst_point = [];
            for boxes in box.split("'"):
                for point in boxes.split(","):
                    lst_point.append(point)
            start = 1;
            step = len(lst_point)
            for i in range(0, round(step/6)):
                    xmin = lst_point[start]
                    ymin = lst_point[start+1]
                    xmax = lst_point[start+2]
                    ymax = lst_point[start+3]
                    start = start+6
                    b = (float(xmin), float(xmax), float(ymin), float(ymax))
                    bb = convert_coordinates((width,height), b)
                    f.write("0" + " " + " ".join([("%.6f" % a) for a in bb]) + '\n')
        print ("wrote %s" % fname_out)
```

```python
cover_toyolo()
```
